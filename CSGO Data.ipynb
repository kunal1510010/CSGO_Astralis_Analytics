```python
data_astralis = data[data['home_team_name'] == 'Astralis']
```

```python
player_name = ['dupreeh', 'device', 'Xyp9x','gla1ve', 'Magisk']
```

```python
player_data = pd.DataFrame(columns = ['Against','Name', 'Kills', 'Headshots', 'Asists', 'Deaths', 'Rating'])
```

```python
new_data_columns = ['Against','Name', 'Kills', 'Headshots', 'Asists', 'Deaths', 'Rating']
```

```python
columns = ['away_team_name','ht_p1_name', 'ht_p1_kills', 'ht_p1_headshots', 'ht_p1_asists', 'ht_p1_deaths', 'ht_p1_rating20']
player_data = pd.DataFrame(columns = ['Against','Name', 'Kills', 'Headshots', 'Asists', 'Deaths', 'Rating'])
col = 'ht_p1_name'
for i in player_name:
    for j in range(1,6):
        current_col = col.replace('1',str(j))
        columns_temp = [ i.replace('1', str(j)) for i in columns]
        data_astralis_temp = data_astralis[data_astralis[current_col] == i][columns_temp]
        data_astralis_temp = data_astralis_temp[columns_temp]
        data_astralis_temp.columns = new_data_columns
        player_data = pd.concat([player_data, data_astralis_temp])
        
```

```python
player_data[player_data['Kills'] == max(player_data['Kills'])]
```

```python
player_data = player_data.astype({'Kills': 'int32','Headshots': 'int32','Asists': 'int32','Deaths': 'int32'})
```

```python
for i in player_name:
    data_temp = player_data[player_data['Name'] == i]
    data_temp = data_temp.sort_values(by = 'Kills', ascending = False)
    print(i)
    print('\n \n')
    print(data_temp.head())
    print('\n \n')
    print(data_temp.describe())
    print('\n \n')
```
